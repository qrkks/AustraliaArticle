## 3 肺癌患者生活质量测定量表 FACT_L (Functional Assessment of Cancer Therapy -Lung)

```python
factl = pd.read_csv('../db/fact_l.csv', encoding='gbk', header=[0,1])
factl.head()
```

```python
## 设置双层索引

factl.columns.names = ['level_0', 'level_1']

# 初始化新列名列表
new_columns = []

# 初始化一个临时变量 previous_name
previous_name = None

# 遍历每一列，检查是否包含 "Unnamed"，如果包含就替换为左侧的列名
for level_0, level_1 in factl.columns:
    if 'Unnamed' in level_0:
        # 处理 'Unnamed' 列，如果是第一列也为 'Unnamed'，则不能找到前面有效的列名
        if previous_name is not None:
            new_columns.append((previous_name, level_1))
        else:
            # 如果第一列是 'Unnamed'，就保持原样
            new_columns.append((level_0, level_1))
    else:
        new_columns.append((level_0, level_1))
        previous_name = level_0

# 将新的列名应用到 factl DataFrame 中
factl.columns = pd.MultiIndex.from_tuples(
    new_columns, names=['level_0', 'level_1'])


factl.head()
```

```python
## 设置索引并将内容里的对象字样去掉，只保留数字

# 去掉列名的 "Unnamed:" 并保留原始列层级
factl.columns.names = ['level_0', 'level_1']

# 提取当前的索引列（这列有 "对象1", "对象2" 的值）
index_column = factl['Unnamed: 0_level_0']['Unnamed: 0_level_1']

# 去掉 "对象" 只保留数字部分
factl.index = index_column.str.replace('对象', '').astype(int)

# 删除原索引列
factl = factl.drop(columns=[('Unnamed: 0_level_0', 'Unnamed: 0_level_1')])

factl.head()
```

```python
## 命名索引
factl.index.name = '对象'
factl.head()
```

```python
## 去除二层索引的unnamed部分

# 获取列的多重索引
columns = factl.columns

# 创建新的多重索引，去掉第二层中的 'Unnamed' 部分
new_columns = []

for level_0, level_1 in columns:
    if 'Unnamed' in level_1:
        new_columns.append((level_0, ''))  # 将 'Unnamed' 替换为空字符串
    else:
        new_columns.append((level_0, level_1))

# 将新的列索引应用到 DataFrame 中
factl.columns = pd.MultiIndex.from_tuples(new_columns, names=['level_0', 'level_1'])

# 查看修改后的 DataFrame
factl.head()
```

```python
# 获取现有的列索引
columns = factl.columns

# 修改第一层索引，将英文括号替换为中文括号
new_columns = []

for level_0, level_1 in columns:
    # 替换 level_0 中的英文括号为中文括号
    new_level_0 = level_0.replace('(', '（').replace(')', '）')
    new_columns.append((new_level_0, level_1))

# 重新设置修改后的多重索引
factl.columns = pd.MultiIndex.from_tuples(new_columns, names=['level_0', 'level_1'])

# 查看修改后的 DataFrame
factl.head()
```

```python
factl.columns.levels
```

```python
factl.columns = factl.columns.set_levels([level.fillna('0') for level in factl.columns.levels])
```

```python
# 使用短横线作为占位符
new_columns = []
for level_0, level_1 in factl.columns:
    if level_1 == '':
        level_1 = '-'  # 使用短横线
    new_columns.append((level_0, level_1))

factl.columns = pd.MultiIndex.from_tuples(new_columns)

print("使用短横线替换后的 DataFrame：")
print(factl)
```

```python
factl.info()
```

```python
factl.to_csv('../db/purged/fact_l.csv',encoding='utf-8-sig')
```
