```python
import matplotlib.pyplot as plt

# 解决中文字体显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置为黑体以支持中文
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题

# 对数据进行排序，按原值排序而不是按频数
value_counts_sorted = data['经济毒性分级'].value_counts().sort_index()

# 绘制条形图，按原值顺序显示
value_counts_sorted.plot(kind='bar')

# 显示图形
plt.xlabel('经济毒性分级')  # 添加x轴标签
plt.ylabel('数量')        # 添加y轴标签
plt.title('经济毒性分级分布图')  # 添加标题
plt.show()
```
